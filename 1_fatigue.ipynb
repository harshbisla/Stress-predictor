## WEIGHTED AVERAGE TECHNIQUE

## Giving weights to questiions and finding the output unified quantity


```python
# Define the weightage for each row 
row_weights_UTC = [0.4, 0.12, 0.2, 0.2, 0.08]
row_weights_non_UTC = [0.4347, 0.13043, 0.21739, 0.21739]

# Define the number of rows in each group
num_rows = len(data)
# Initialize a list to store the calculated values
calculated_values = []

# Initialize an index variable
i = 0
# Iterate over rows in groups
while i < num_rows:
    
    # Get the current group of rows
    if data.iloc[i]["Timezone"]=="UTC":
        group = data.iloc[i:i+5]
        weighted_average = (group["q_ans"] * row_weights_UTC[:len(group)]).sum()
        subject_id = group["SubjectID"].iloc[0]
        datetime = group["DateTime"].iloc[0]
        calculated_values.append({"Subject_ID": subject_id, "DateTime": datetime, "Average_Stress": weighted_average})
        
        # calculated_values.append(weighted_average)  
        i+=5
    else:
        group = data.iloc[i:i+4]
        weighted_average = (group["q_ans"] * row_weights_non_UTC[:len(group)]).sum()
        subject_id = group["SubjectID"].iloc[0]
        datetime = group["DateTime"].iloc[0]
        calculated_values.append({"Subject_ID": subject_id, "DateTime": datetime, "Average_Stress": weighted_average})
        # calculated_values.append(weighted_average)     
        i+=4

# Perform further calculations or analysis on the calculated values as needed
print("Calculated Values:", calculated_values)
len(calculated_values)
```

```python
# Convert calculated_values to a DataFrame
df = pd.DataFrame(calculated_values)

# Print the resulting DataFrame
df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=45)
```

## Predicting

```python
import pickle

# Save the model as a pickle file
with open("trained_model.pkl", "wb") as file:
    pickle.dump(model, file)
```

```python
questions = {
    "Q1": "Describe fatigue on a scale of 1 to 10, where 1 means you don’t feel tired at all and 10 means the worst tiredness you can imagine: ",
    "Q2": "Are you feeling better, worse, or the same as yesterday?: ",
    "Q3": "Physically, today how often did you feel exhausted? (never, sometimes,often ,regularly ,always)",
    "Q4": "Mentally, today how often did you feel exhausted? (never, sometimes,often ,regularly ,always): ",
    "Q5": "Did you do sport today? (Yes/No): "
}
```

```python
mapping = {
    
    "Better": 0,
    "Same": 1,
    "Worse": 2,
    "Never":0,
    "Sometimes":1,
    "Often":2,
    "Regularly":3,
    "Always":4,
    "Yes": 1,
    "No": 0
}
```

```python
user_data = []

# Ask the user for answers to the PRO questions
for key, question in questions.items():
    answer = input(question)
    if isinstance(answer, str):  # Check if the answer is a string
        mapped_answer = mapping.get(answer, None)  # Map the answer using the mapping dictionary
        if mapped_answer is not None:
            user_data.append(mapped_answer)
        else:
            user_data.append(int(answer))
    else:
        user_data.append(int(answer))

print(user_data)
```

```python

# Normalize the responses
num_inputs = [10,2,4,4,1]
result = [u / n for u, n in zip(user_data, num_inputs)]

#Finding Weighted average
weights = [0.3, 0.175, 0.175, 0.175, 0.175]
weighted_stress = sum(r * w for r, w in zip(result, weights))
print(weighted_stress)

#Load the trained model
with open('trained_model.pkl', 'rb') as file:
    model = pickle.load(file)


# Create a DataFrame with the input data
input_data = pd.DataFrame({'Average_Stress': [weighted_stress]})


# Make predictions using the loaded model
predictions = model.predict(input_data)


# Print the prediction
if predictions == 1:
    print("Based on your answers, you are classified as stressed.")
else:
    print("Based on your answers, you are classified as not stressed.")
```
