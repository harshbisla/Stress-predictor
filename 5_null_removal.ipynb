## Making number of rows equal in each file

```python
import os
import pandas as pd
import numpy as np
```

```python
folder_path = 'updated_data_sliding_norm_dated'
# output_folder = 'data_balanced'
output_folder = 'updated_data_sliding_norm_dated'
```

to find maximum no of rows in file

```python
# Folder path where CSV files are located
folder_path = 'updated_data_sliding_norm_dated'

# Get a list of all files in the folder
file_list = os.listdir(folder_path)

# Initialize variables to keep track of maximum rows and the file with maximum rows
max_rows = 0
file_with_max_rows = ""

# Iterate through each file in the folder
for file_name in file_list:
    file_path = os.path.join(folder_path, file_name)
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(file_path)
    # Count the number of rows in the DataFrame
    num_rows = len(df)
    print(f"File: {file_name}, Rows: {num_rows}")
    
    # Update the maximum rows and file_with_max_rows if necessary
    if num_rows > max_rows:
        max_rows = num_rows
        file_with_max_rows = file_name

print(f"File with the maximum number of rows: {file_with_max_rows}, Rows: {max_rows}")
```

```python
# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Get a list of all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]


for file in csv_files:
    file_path = os.path.join(folder_path, file)

    print(f"Processing file: {file}")
     
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(file_path)


    # Rename the first column as "DateTime"
    df = df.rename(columns={df.columns[0]: "DateTime"})
    
    
    # Drop the 'DateTime' column
    df = df.drop('DateTime', axis=1)

    # Check the number of rows in the DataFrame
    num_rows = len(df)

    # If the number of rows is less than 288, create new rows and fill them with values from the existing rows
    if num_rows < 720:
        # if num_rows < 10:
        #     continue
        
        num_new_rows = 720 - num_rows
        
        # Generate new rows by repeating the existing rows
        num_repetitions = (num_new_rows // num_rows) + 1
        new_rows = np.tile(df.values, (num_repetitions, 1))[:num_new_rows]
        

        # Append the new rows to the DataFrame
        df = pd.concat([df, pd.DataFrame(new_rows, columns=df.columns)], ignore_index=True)
        num_rows = len(df)

    # Save the modified DataFrame to a new CSV file in the output folder
    output_file_path = os.path.join(output_folder, file)
    df.to_csv(output_file_path, index=False)
```
